```python
import matplotlib.pyplot as plt
import math
```

```python
# Function to return the student scores and total number of students
def mainFunction():
    student_scores = [11,14,4,18,3,8,8,5,14,7,2,4,7,9,4,6,10,2,10,16,5,0,19,17,20,13,
                      1,5,19,16,15,2,19,8,10,8,12,15,13,6,2,7,4,19,17,6,7,5,15,5,13,
                      2,15,4,15,1,10,18,1,7,9,6,13,15,9,20,7,15,20,12,3,0,15,17,3,0,
                      18,20,19,2,11,11,19,17,16,18,15,0,9,17,6,12,3,4,8,17,16,17,4,1]
    total_students = len(student_scores)
    return student_scores, total_students

# Function to calculate and return the mean (average) of the scores
def calculateMean(student_scores, total_students):
    total_sum = sum(student_scores)  # Find the total sum of all scores
    mean = total_sum / total_students  # Divide total by number of students
    return mean

# Function to calculate and return the median
def calculateMedian(student_scores, total_students):
    sorted_scores = sorted(student_scores)  # Sort scores in ascending order
    middle = total_students // 2

    if total_students % 2 == 0:
        # If even number of students, average the two middle scores
        median = (sorted_scores[middle - 1] + sorted_scores[middle]) / 2
    else:
        # If odd, take the middle score
        median = sorted_scores[middle]
    return median

# Function to calculate and return the mode(s)
def calculateMode(student_scores):
    frequency_dict = {}  # Dictionary to count frequency of each score

    # Count how many times each score appears
    for score in student_scores:
        frequency_dict[score] = frequency_dict.get(score, 0) + 1

    # Find the maximum frequency
    max_frequency = max(frequency_dict.values())

    # Return all scores that appear with that frequency
    mode = [score for score, freq in frequency_dict.items() if freq == max_frequency]
    return mode

# Function to calculate and return the standard deviation
def standardDev(student_scores, mean, total_students):
    sum_sqr_diff = 0
    for score in student_scores:
        # Calculate squared difference between each score and the mean
        sqr_diff = (score - mean) ** 2
        sum_sqr_diff += sqr_diff  # Add to the total sum of squared differences

    # Divide by total and take square root to get standard deviation
    deviation = math.sqrt(sum_sqr_diff / total_students)
    return deviation

# Main function to run all calculations and show histogram
def main():
    student_scores, total_students = mainFunction()

    # Calculate statistics
    mean = calculateMean(student_scores, total_students)
    median = calculateMedian(student_scores, total_students)
    mode = calculateMode(student_scores)
    deviation = standardDev(student_scores, mean, total_students)

    # Print results
    print("\n--- Summary Statistics ---")
    print("Mean:", round(mean, 2))
    print("Median:", round(median, 2))
    print("Mode:", mode)
    print("Standard Deviation:", round(deviation, 2))

    # --- VISUALISATION ---
    # Create a histogram to show distribution of scores
    plt.hist(student_scores, bins=21, color='lightblue', edgecolor='grey')

    # Add lines for mean, median, mode, and standard deviation
    plt.axvline(mean, linewidth=1, label="Mean", color='red')
    plt.axvline(median, linestyle='dashed', linewidth=1, label="Median", color='green')
    for m in mode:
        plt.axvline(m, linestyle='dashed', linewidth=1, label="Mode", color='purple')
    plt.axvline(deviation, linestyle='dashed', linewidth=1, label="Std Deviation", color='blue')

    # Add labels and legend
    plt.title("Test Scores")
    plt.xlabel("Student Scores")
    plt.ylabel("Number of Students")
    plt.grid(True)
    plt.legend()
    plt.show()

# Run the main function
main()
```
